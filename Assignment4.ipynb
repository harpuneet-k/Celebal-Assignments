```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
titanic= sns.load_dataset('titanic')

# Display the first few rows
titanic.head()
```

```python
# Basic info
print(titanic.info())

# Statistical summary
print(titanic.describe(include='all'))

# Missing values
print(titanic.isnull().sum())
```

```python
#HANDLING MISSING VALUES
# Fill 'age' with median, 'embarked' with mode, drop 'deck' due to many missing values
titanic['age'].fillna(titanic['age'].median(), inplace=True)
titanic['embarked'].fillna(titanic['embarked'].mode()[0], inplace=True)
titanic.drop(columns=['deck'], inplace=True)
```

```python
#UNIVARIATE ANALYSIS
#Survival Count
sns.countplot(data=titanic, x='survived')
plt.title("Survival Count")
plt.show()
```

```python
#Class Distribution
sns.countplot(data=titanic, x='pclass',color='green')
plt.title("Passenger Class Distribution")
plt.show()
```

```python
#Age Distribution
plt.hist(titanic['age'], bins=30, color='pink', edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

```

```python
#BIVARIATE ANALYSIS
#Survival by Gender
sns.countplot(data=titanic, x='sex', hue='survived')
plt.title("Survival Count by Gender")
plt.show()
```

```python
#Survival by class
sns.countplot(data=titanic, x='pclass', hue='survived', color='pink')
plt.title("Survival by Passenger Class")
plt.show()
```

```python
  #Age vs Survival
  sns.histplot(data=titanic, x='age', hue='survived', multiple='stack')
plt.title("Age vs Survival")
plt.show()

```

```python
#CORRELATION MATRIX AND HEATMAP
corr_matrix = titanic.corr(numeric_only=True)

plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True, cmap='PiYG', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
#PAIRPLOT FOR KEY FEATURES
sns.pairplot(titanic[['survived', 'age', 'fare', 'pclass']], hue='survived')
plt.show()
```

```python
#BOXPLOT OF AGE BY CLASS AND GENDER
sns.boxplot(data=titanic, x='pclass', y='age', hue='sex',color='green')
plt.title("Age Distribution by Class and Gender")
plt.show()
```

```python
#FARE DISTRIBUTION BY CLASS
sns.violinplot(data=titanic, x='pclass', y='fare', hue='survived', split=True)
plt.title("Fare Distribution by Class and Survival")
plt.show()
```
